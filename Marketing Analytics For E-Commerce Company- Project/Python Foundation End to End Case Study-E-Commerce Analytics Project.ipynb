### 3. Cross-Selling (Which products are selling together)
### Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)

```python
import itertools

# Group the data by order_id and aggregate with product names as a list
order_product_combinations = merged_df.groupby('order_id')['product_id'].agg(list).reset_index()

# Create an empty dictionary to store combinations and their frequencies
combination_counts = {}

# Iterate over each row in the order_product_combinations DataFrame
for products in order_product_combinations['product_id']:
    # Generate all combinations of 2 and 3 products
    for combination_length in range(2, 4):
        for combination in itertools.combinations(products, combination_length):
            # Sort the combination to ensure consistency in counting
            sorted_combination = tuple(sorted(combination))
            # Increment the count for the combination
            combination_counts[sorted_combination] = combination_counts.get(sorted_combination, 0) + 1

# Sort the combinations by frequency in descending order
sorted_combinations = sorted(combination_counts.items(), key=lambda x: x[1], reverse=True)

# Select the top 10 combinations
top_10_combinations = sorted_combinations[:10]

# Print the top 10 combinations
print("Top 10 Combinations of Products Selling Together:")
for combination, count in top_10_combinations:
    print(f"Combination: {combination}, Count: {count}")
```
