```python
# Filter out UK entries
non_uk_data = online_rt[online_rt['Country'] != 'UK']

# Group by Country and sum the Quantity for each country
country_quantity = non_uk_data.groupby('Country')['Quantity'].sum().nlargest(10)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
country_quantity.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Quantity Ordered (Excluding UK)')
plt.xlabel('Country')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.tight_layout()
plt.show()
```

### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries
Hint: First we need to find top-3 countries based on revenue, then create scater plot between Quantity and Unitprice for each country separately


```python
# revenue (Quantity * UnitPrice) for each row
online_rt['Revenue'] = online_rt['Quantity'] * online_rt['UnitPrice']

# Group by Country and total revenue for each country
country_revenue = online_rt.groupby('Country')['Revenue'].sum().nlargest(3)

# top 3 countries based on revenue
top_3_countries = country_revenue.index.tolist()

# Filter data for the top 3 countries
top_3_data = online_rt[online_rt['Country'].isin(top_3_countries)]

# scatter plots for each country separately
plt.figure(figsize=(10, 6))

for country in top_3_countries:
    country_data = top_3_data[top_3_data['Country'] == country]
    plt.scatter(country_data['Quantity'], country_data['UnitPrice'], label=country)

plt.title('Quantity per UnitPrice by CustomerID for Top 3 Countries (Based on Revenue)')
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 86. Create Stacked Bar Chart for Soap, Shampo, ToothPaste for each month
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
months = company_data['Month']
soap_sales = company_data['Soap']
shampoo_sales = company_data['Shampo']
toothpaste_sales = company_data['ToothPaste']

# Creating the stacked bar chart
plt.figure(figsize=(10, 6))  
bar_width = 0.35
index = range(len(months))

# Plotting stacked bars
plt.bar(index, soap_sales, width=bar_width, label='Soap')
plt.bar(index, shampoo_sales, width=bar_width, label='Shampo', bottom=soap_sales)
plt.bar(index, toothpaste_sales, width=bar_width, label='ToothPaste',
        bottom=[soap + shampoo for soap, shampoo in zip(soap_sales, shampoo_sales)])

plt.xlabel('Month')
plt.ylabel('Number of Units Sold')
plt.title('Units Sold per Month for Soap, Shampo, ToothPaste')
plt.xticks(index, months, rotation=45)  
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
